```python
from skimage import data, feature
import matplotlib.pyplot as plt
```

```python
image = data.astronaut()
print(image.shape)
```

```python
hog_vector, hog_image = feature.hog(image, orientations=9, pixels_per_cell=(8,8), cells_per_block=(2,2), block_norm='L2', visualize=True, channel_axis=2)
```

```python
figure, axes = plt.subplots(1, 2, figsize=(12, 6), subplot_kw=dict(xticks=[], yticks=[]))
axes[0].imshow(image)
axes[0].set_title('Original Image')

axes[1].imshow(hog_image)
axes[1].set_title('HOG Image')
plt.show()
```
