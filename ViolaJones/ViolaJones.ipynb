```python
import cv2
import matplotlib.pyplot as plt
```

```python
cascade_classifier = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
image = cv2.imread('face.jpeg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
detected_faces = cascade_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=10)
```

```python
for (x, y, width, height) in detected_faces:
    cv2.rectangle(image, (x, y), (x + width, y + height), (0, 0, 255), 10)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
mul_image = cv2.imread('people_face.jpeg')
gray_mul_image = cv2.cvtColor(mul_image, cv2.COLOR_BGR2GRAY)
detected_faces = cascade_classifier.detectMultiScale(gray_mul_image, scaleFactor=1.1, minNeighbors=3, minSize=(20, 20))

for (x, y, width, height) in detected_faces:
    cv2.rectangle(mul_image, (x, y), (x + width, y + height), (0, 0, 255), 3)

plt.imshow(cv2.cvtColor(mul_image, cv2.COLOR_BGR2RGB))
plt.show()
```
