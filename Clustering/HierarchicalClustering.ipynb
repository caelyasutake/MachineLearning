```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as cluster_algorithm
from sklearn.cluster import AgglomerativeClustering
```

```python
shopping_data = pd.read_csv("Mall_Customers.csv")
data = shopping_data.iloc[:, 3:5].values
```

```python
plt.figure(figsize=(10,7))
plt.title("Market Segmentation Dendogram")
dendrogram = cluster_algorithm.dendrogram(cluster_algorithm.linkage(data, 'ward'))
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:, 0], data[:, 1], c=cluster.labels_, cmap='rainbow')
```
